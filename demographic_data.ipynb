```python
import requests
import json
```

```python
url = "https://bdl.stat.gov.pl/api/v1/Variables?subject-id=P2425"
response = requests.get(url)
```

```python
if response.status_code == 200:
    data = response.json()['results']
    print(json.dumps(data, indent=4))
```
