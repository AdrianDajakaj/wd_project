```python
df_1 = pd.read_csv('measures_history/białystok-miejska-air-quality.csv')
```

```python
df_1['date'] = pd.to_datetime(df_1['date'], format='%Y/%m/%d')
df_1['date'] = pd.to_datetime(df_1['date'].dt.strftime('%Y-%m-%d'))
df_sorted = df_1.sort_values(by='date', ascending=False)
```

```python
to_float_columns = df_sorted.columns.difference(['date'])
df_sorted[to_float_columns] = df_sorted[to_float_columns].applymap(lambda x: pd.to_numeric(x, errors='coerce'))
```

```python
df_sorted.head()
```

```python
df_sorted.info()
```

```python
df_sorted['year'] = df_sorted['date'].dt.year
df_sorted['month'] = df_sorted['date'].dt.month

# Grupowanie po roku i miesiącu, obliczanie średniej dla każdej kolumny numerycznej
monthly_avg = df_sorted.groupby(['year', 'month']).mean().reset_index()

# Przygotuj nowy DataFrame dla wyników
result_df = pd.DataFrame()

# Dodaj kolumny 'year' i 'month' do nowego DataFrame
result_df['year'] = monthly_avg['year']
result_df['month'] = monthly_avg['month']

num_cols = [col for col in monthly_avg.columns if col not in ['year', 'month', 'date']]
for col in num_cols:
    result_df[col] = monthly_avg[col]
result_df = result_df.sort_values(by=['year', 'month'], ascending=False)
```
